```python
## using inline plots
%matplotlib inline

## importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
#import pylab as py
import ap21_module as main
import uncertainties as unc
import uncertainties.unumpy as unp
from scipy.optimize import curve_fit
#from scipy.stats import chi2
```

1) Quantitative Beobachtungen am Einzelspalt
===

```python
### werte für plot von Lage der minima vs. ordnung (neg. ordung für positive steigung)
order = np.array([-5, -4, -3, -2, -1, 0])
pos_min_px = np.array([544.95, 583.06, 622.94, 663.42, 702.83]) # position der mimina in pixels
pos_max_px = np.array([526.38, 565.47, 605.53, 643.45, 686.64, 746.02]) # pos. der max. in pix.

### linear fit:
def lin_reg(x, k, d):
    return k*x+d

popt, pcov = curve_fit(lin_reg, order[:5], pos_min_px)
popt_err = np.sqrt(np.diag(pcov))
print('optimale parameter u error:', popt, popt_err, '\n')

### parameter-string für plot:
par_vals = r'$k=39.61\pm0.23\ \mathrm{[px/Ordnung]}$' + '\n' + r'$d=742.28\pm0.75\ \mathrm{[px]}$'

### Ordnung der Maxima (berechnet)
print('Berechnete Ordnung der Maxima')
for i in pos_max_px:
    print(i, (i-popt[1])/popt[0])
    
### plotting section
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.errorbar(order[:5], pos_min_px, yerr=10, fmt='r.', label='$Messwerte:\ Lage\ der\ Minima$')
ax.plot(np.linspace(-6, 0.5, 10), lin_reg(np.linspace(-6, 0.5, 10), *popt), 'k--', label='$Linearer\ Fit:\ k*x+d$')
ax.errorbar((pos_max_px-popt[1])/popt[0], pos_max_px, yerr=10, fmt='k.', label='$Messwerte:\ Lage\ der\ Maxima$')
ax.set_xlim(-6, 0.5)
ax.set_title('$Plot\ 1:\ Lage\ der\ Maxima/Minima\ vs.\ Ordnung$')
ax.set_ylabel('$Lage\ \mathrm{[px]}$')
ax.set_xlabel('$Ordnung$')
ax.text(-5.9, 730, par_vals)
ax.grid(axis='both')
ax.legend(loc=2)
#plt.savefig('.data/plot1.pdf', dpi=200, format='pdf')
```
