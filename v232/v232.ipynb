1) Wellenlänge des grünen Lasers
===

```python
## values
s0 = np.array([0.000, 0.030, 0.100, 0.200, 0.300]) # in mm
se = np.array([2.966, 2.987, 3.066, 3.166, 3.267]) # in mm
s = unp.uarray(np.mean(se-s0), 9e-3/np.sqrt(5)) # Gesamtweg mit Fehler
N = np.array([11167, 11136, 11169, 11172, 11174])
N = unp.uarray(np.mean(N), 10)
lamb = 2 * s / N * 1e-3
print('Wellenlänge:', 2 * s / N * 1e6,'nm')
print('Weg s:',s,'mm')
print('Zählrate:',N)
```

2) Brechungsindex von Luft
===

```python
## values (3 measurements), 1 torr = 101325/760 Pa 
## delta_m is always 5, except for the first (0th) value

p1 = unp.uarray([-750, -685, -610, -540, -460, -385, -310, -230, -160, -80, -10], 5)
p2 = unp.uarray([-750, -685, -610, -530, -455, -385, -310, -235, -155, -80, -10], 5)
p3 = unp.uarray([-755, -685, -610, -530, -460, -385, -305, -235, -160, -80, -10], 5)
p = (p1+p2+p3)/3 # mean
p = p * 101325/760 ## Umrechnung auf Pa

delta_m = np.arange(0, 55, 5)

T = unc.ufloat(273.15+23.8, 0.1)


### Brechungsindex
p0 = 101325 ## [Pa], Normaldruck
T0 = 273.15 ## [K], Normaltemp
a = unc.ufloat(50e-3, 0.05e-3) ## Innenmaß der Küvette, [m]
#lamb = unc.ufloat(532e-9, 1e-9)

#n0 = (lamb*delta_m*p0*T) / (2*a*p*T0) + 1
#print('n0 =', np.mean(n0))

### linear fit
def lin_reg(x, k, d):
    return k*x+d

popt, pcov = curve_fit(lin_reg, unp.nominal_values(p), delta_m)
popt_err = np.sqrt(np.diag(pcov))
print(popt, popt_err)
k = unc.ufloat(popt[0], popt_err[0])
### fit-values string
fit_vals = r'$k=(5.01\pm0.02)\cdot10^{-4}\ \mathrm{[Streifen/Pa]}$' + "\n" + r'$d=50.58\pm0.10\ \mathrm{[Streifen]}$'

### plotting
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.errorbar(unp.nominal_values(p), delta_m, xerr=unp.std_devs(p), fmt='r.', label='$Messwerte\ Druck\ p$')
ax.plot(unp.nominal_values(p), lin_reg(unp.nominal_values(p), *popt), 'k--', label='$Linearer\ Fit:\ k\cdot p + d$')
ax.text(-118500, 43, fit_vals)
ax.set_ylim(-2, 52)
ax.set_title('$Plot\ 1:\ Interferenzstreifen\ \\vartriangle m\ vs.\ Luftdruck\ p$')
ax.set_ylabel('$Anzahl\ gezählter\ Interferenzstreifen\ \\vartriangle m$')
ax.set_xlabel('$Luftdruck\ p\ (relativ\ zu\ p_0)\ \\mathrm{[Pa]}$')
ax.grid(axis='both')
ax.legend(loc=0)
#plt.savefig('.data/plot1.pdf', dpi=200, format='pdf')
```

```python
n0 = (lamb*k*p0*T) / (2*a*T0) + 1
print(n0)
```

3) Kohärenzlänge der Leuchtdiode
===

```python
### values
t, U = np.loadtxt('.data/NewFile0.csv', delimiter=',', usecols=(0,1), skiprows=2, unpack=True)

### Gauss-fit
def gaussian(x, a, mu, sig):
    return a * np.exp(-(x-mu)**2/(2*sig**2))

# import signal from scipy (bad coding style, I know.)
from scipy import signal

ind = signal.find_peaks_cwt(U, np.arange(2,2.001)) # find the indices of the peaks in U
U_new = [[U[i],i] for i in ind if U[i]>0.0] # U_new is an array of [U,index] pairs, where U is greater than 0.
U_new_ind = [i[1] for i in U_new] # U_new_ind are the indices of the previously found U_new's.
# These indices are used for the curve_fit, to find parameters for the gaussian fit. 

start_vals = [0.05, 0.02, 0.04]
popt, pcov = curve_fit(gaussian, t[U_new_ind], U[U_new_ind], p0=start_vals)
popt_err = np.sqrt(np.diag(pcov))
print(popt, popt_err)
# the values of curve_fit for mu and sig are good, just the amplitude is a bit too low. changed that by hand. ;-)

### fit_values string
fit_vals = r'$A=0.047\pm0.007\ \mathrm{[V]}$' + "\n" + r'$\mu=0.021\pm0.001\ \mathrm{[s]}$' + "\n" + r'$\sigma=0.015\pm0.001\ \mathrm{[s]}$'

### plotting
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.plot(t, U, 'r-', label='$Messwerte$')
ax.plot(t, gaussian(t, 0.047, 0.0213, 0.0149), 'k--', label='$Gauß-Fit:\ A \cdot \mathrm{e}^{\\frac{-(t-\mu)^2}{2\sigma^2}}$')
ax.set_title('$Plot\ 2:\ Signalverlauf\ des\ Wellenpakets\ mit\ Gauß-Fit$')
ax.set_xlabel('$Zeit\ t\ \mathrm{[s]}$')
ax.set_ylabel('$Amplitude\ \mathrm{[V]}$')
ax.text(0.095, 0.027, fit_vals)
ax.grid(axis='both')
ax.legend(loc=t0, fontsize=14)
#plt.savefig('.data/plot2.pdf', dpi=200, format='pdf')
```
