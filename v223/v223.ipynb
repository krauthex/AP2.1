2) Histogramm von $\vartriangle$x, $\vartriangle$y
===

```python
hist_array = sorted(np.hstack((dx,dy)))
hist_mean = np.mean(hist_array)
hist_std = np.std(hist_array)
print('mu=',hist_mean,'sig=',hist_std)

###gauss-fit für die bestimmten werte
def gauss(x, mu, sig):
    return np.exp(-((x-mu)**2)/(2*sig**2))/np.sqrt(2*np.pi*sig**2)

###D, kB aus sigma
eta = unc.ufloat(9.4e-4,0.05e-4)
a = unc.ufloat(755e-9, 30e-9)
T = unc.ufloat(273.15+22.8, 0.1)
D = (hist_std * 1e-6)**2 / 4
k = D * 6*np.pi*eta*a / T
print(k, D)

#### plotting
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.hist(hist_array, bins=8, color='r', normed=True, label='$\\vartriangle x,\ \\vartriangle y$')
ax.plot(hist_array, gauss(hist_array, hist_mean, hist_std), 'k--', 
        label='$Gauß-Fit:\ \\frac{1}{\sqrt{2\pi\sigma^2}}\cdot\mathrm{e}^{\\frac{-(x-\mu)^2}{2\pi\sigma^2}}$')
ax.set_title('$Plot\ 2:\ Histogramm\ von\ \\vartriangle x, \\vartriangle y\ mit\ Gauß-Fit$')
ax.set_xlabel('$\\vartriangle x, \\vartriangle y\ \mathrm{[\mu m]}$')
ax.set_ylabel('$Normierte\ Häufigkeit/\ Gauß-Fit$')
ax.grid(axis='both')
ax.text(2.2, 0.32, r'$\mu=0.0104\ \mathrm{[\mu m]}$' + "\n" + r'$\sigma=1.0821\ \mathrm{[\mu m]}$')
ax.legend(loc=0)
#plt.savefig('.data/plot2.pdf', dpi=200, format='pdf')
```

3) Kummulative Verteilung der Verschiebungsquadrate
===

```python
### linear fit

def lin_reg(x, k, d):
    return k*x+d

popt, pcov = curve_fit(lin_reg, t[:149], np.cumsum(r2[:149]))
popt_err = np.sqrt(np.diag(pcov))

D = unc.ufloat(popt[0], popt_err[0])
print(popt, popt_err)
D = D/4 * 1e-12
print(D)
```

```python
###value-string
values = r'$k=2.30\pm0.02\ \mathrm{[\frac{(\mu m)^2}{s}]}$' + "\n" + r'$d=2.93\pm1.45\ \mathrm{[(\mu m)^2]}$'

###plotting
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.plot(t[:149], np.cumsum(r2[:149]), 'r.', label='$\Sigma\ r^2$')
ax.plot(t[:149], lin_reg(t[:149], *popt), 'k--', label='$Linearer\ Fit:\ k \cdot t+d$')
ax.set_title('$Plot\ 3:\ Kummulative\ Verschiebung\ eines\ Partikels\ mit\ linearem\ Fit$')
ax.set_xlabel('$Zeit\ t\ \mathrm{[s]}$')
ax.set_ylabel('$Summe\ der\ Verschiebequadrate\ \Sigma\ r^2\ \mathrm{[(\mu m)^2]}$')
ax.text(2, 320, values, fontsize=11 )
ax.grid(axis='both')
ax.legend(loc=2)
plt.savefig('.data/plot3.pdf', dpi=200, format='pdf')
```
