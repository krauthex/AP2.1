2) Histogramm von $\vartriangle$x, $\vartriangle$y
===

```python
hist_array = sorted(np.hstack((dx,dy)))
hist_mean = np.mean(hist_array)
hist_std = np.std(hist_array)


def gauss(x, mu, sig):
    return np.exp(-((x-mu)**2)/(2*sig**2))/np.sqrt(2*np.pi*sig**2)

#### plotting

fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.hist(hist_array, bins=8, color='r', normed=True, label='$\\vartriangle x,\ \\vartriangle y$')
ax.plot(hist_array, gauss(hist_array, hist_mean, hist_std), 'k--', 
        label='$Gauß-Fit:\ \\frac{1}{\sqrt{2\pi\sigma^2}}\cdot\mathrm{e}^{\\frac{-(x-\mu)^2}{2\pi\sigma^2}}$')
ax.set_title('$Plot\ 2:\ Histogramm\ von\ \\vartriangle x, \\vartriangle y\ mit\ Gauß-Fit$')
ax.set_xlabel('$\\vartriangle x, \\vartriangle y\ \mathrm{[\mu m]}$')
ax.set_ylabel('$Normierte\ Häufigkeit/\ Gauß-Fit$')
ax.grid(axis='both')
ax.legend(loc=0)
```

3) Kummulative Verteilung der Verschiebungsquadrate
===

```python
### linear fit

def lin_reg(x, k, d):
    return k*x+d

popt, pcov = curve_fit(lin_reg, t[:149], np.cumsum(r2[:149]))
popt_err = np.sqrt(np.diag(pcov))

print(popt, popt_err)
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.plot(t[:149], np.cumsum(r2[:149]), 'r.', label='$\Sigma\ r^2$')
ax.plot(t[:149], lin_reg(t[:149], *popt), 'k--', label='$Linearer\ Fit:\ k \cdot x+d$')
ax.set_title('$Plot\ 3:\ Kummulative\ Verschiebung\ eines\ Partikels\ mit\ linearem\ Fit$')
ax.set_xlabel('$Zeit\ \mathrm{[s]}$')
ax.set_ylabel('$Summe\ der\ Verschiebequadrate\ \Sigma\ r^2$')
ax.grid(axis='both')
ax.legend(loc=0)
```
